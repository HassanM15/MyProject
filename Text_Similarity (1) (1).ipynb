```python
import pandas as pd
import abydos
import numpy
import collections.abc
#hyper needs the four following aliases to be done manually.
collections.Iterable = collections.abc.Iterable
collections.Mapping = collections.abc.Mapping
collections.MutableSet = collections.abc.MutableSet
collections.MutableMapping = collections.abc.MutableMapping
from abydos.distance import Levenshtein
```

```python
df = pd.read_csv('/content/train.csv',nrows=100)
print(df.head())
```

```python
df = df.drop(['id','qid1','qid2','is_duplicate'],axis=1)
display(df)
```

```python
# list of all abydos distance measures
# https://abydos.readthedocs.io/en/latest/abydos.distance.html#abydos.distance.DamerauLevenshtein
```

# Levenshtein distance

```python
from abydos.distance import Levenshtein
cmp = Levenshtein()
l = []
for i in range(0 , 100):
    l.append(cmp.dist(str(df['question1'][i]),str(df['question2'][i])))
```

```python
l
```

```python
df['Levenshtein'] = l
```

```python
display(df)
```

# Damerau-Levenshtein distance

```python
from abydos.distance import DamerauLevenshtein
cmp = DamerauLevenshtein()
l2 = []
for i in range(0 , 100):
    l2.append(cmp.dist(str(df['question1'][i]),str(df['question2'][i])))
```

```python
l2
```

```python
df['DamerauLevenshtein'] = l2
display(df)
```

```python
from abydos.distance import ShapiraStorerI
cmp = ShapiraStorerI()
l3 = []
for i in range(0 , 100):
    l3.append(cmp.dist(str(df['question1'][i]),str(df['question2'][i])))
```

```python
df['ShapiraStorerI'] = l3
display(df)
```

```python
from abydos.distance import Marking
cmp = Marking()
l4 = []
for i in range(0 , 100):
    l4.append(cmp.dist(str(df['question1'][i]),str(df['question2'][i])))
```

```python
df['Marking'] = l4
display(df)
```

```python
from abydos.distance import MarkingMetric
cmp = MarkingMetric()
l5 = []
for i in range(0 , 100):
    l5.append(cmp.dist(str(df['question1'][i]),str(df['question2'][i])))
```

```python
df['MarkingMetric'] = l5
display(df)
```

```python
df.to_csv('output.csv', index=False)
```
